# jumlah kandungan kolesterol dalam beberapa makanan

```python
models = df.groupby('Shrt_Desc').count()[['Cholestrl']].sort_values(by='Cholestrl',ascending=True).reset_index()
models = models.rename(columns={'Cholestrl':'foodOfColestrol'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Shrt_Desc'], y=models['foodOfColestrol'], color='blue')
plt.xticks(rotation=60)
```

# 4. MODELLING DATA

# seleksi fitur

```python
features = ['Carbohydrt','Calcium','Water','Sodium','Zinc']
x = df[features]
y = df['Cholestrl']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(X_test)
```

# 5. EVALUATION DATA

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
input_data = np.array([[4.78,643,50.01,700,2.79]])

prediction = lr.predict(input_data)
print('Estimasi kandungan kolesterol dalam makanan :', prediction)
```
